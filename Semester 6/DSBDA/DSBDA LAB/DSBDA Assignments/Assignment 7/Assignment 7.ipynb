```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('heart.csv')
```

```python
data
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.fillna(data.mean(), inplace=True)
```

```python
data.isna().sum()
```

# Bar Plot

```python
#Bar plots help visualize the distribution of categorical variables.
#Bar Plot - Count of 'sex' values
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=data, palette='pastel')
plt.xlabel('Sex (0: Female, 1: Male)')
plt.ylabel('Count')
plt.title('Bar Plot - Count of Sex Values')

# Adding a legend
plt.legend(labels=['Female', 'Male'])

plt.show()
```

# Histogram

```python
#Histograms display the distribution of a single variable.
#Histogram - Distribution of 'thalach
plt.figure(figsize=(8,6))
plt.hist(data['thalach'], bins=20, color='skyblue',edgecolor='black')
plt.xlabel('Maximum heart rate')
plt.ylabel('Frequency')
plt.title('Histogram - Distribution of Thalach')
plt.show()
```

```python
#Histogram - Distribution of 'chol'
plt.figure(figsize=(8,6))
plt.hist(data['chol'], bins=20, color='red',edgecolor='black')
plt.xlabel('Cholestrol')
plt.ylabel('Frequency')
plt.title('Histogram - Distribution of Cholestrol')
plt.show()
```

# Scatter Plot

```python
#Visualizing the relationship between two numerical variables.
colors = {0: 'blue', 1: 'orange'}

plt.figure(figsize=(8, 6))
scatter = plt.scatter(data['trestbps'], data['chol'], c=data['sex'].map(colors))

plt.xlabel('Resting blood pressure')
plt.ylabel('Cholesterol')
plt.title('Scatter Plot - Cholesterol vs Resting Blood Pressure')

# Creating a legend
legend_labels = ['Female', 'Male']
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=colors[i], markersize=10) for i in colors.keys()]
plt.legend(legend_handles, legend_labels)

plt.show()
```

# Bar Plot
